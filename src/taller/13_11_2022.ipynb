# MATPLOTLIB


Matplotlib:
 es una biblioteca para la creación y visualización de  gráficos 2D y 3D   en una variedad de formatos impresos.

Ventajas de esta biblioteca:

* Soporte para etiquetas y textos con formato $\LaTeX$
* Control de cada elemento en una figura, incluido el tamaño de la figura y DPI.
* Salida en muchos formatos, incluidos PNG, PDF, SVG, EPS y PGF.
* Todos los aspectos de la figura se pueden controlar *programáticamente*. (Buena reproducibilidad)

Página web de Matplotlib: http://matplotlib.org/

Hay dos enfoques es de vital importancia al trazar con matplotlib, El primero está basado en MATLAB y utiliza una interfaz basada en estados. La segunda opción es una interfaz orientada a objetos.

## La API orientada a objetos matplotlib

En lugar de crear una nueva instancia de figura global, almacenamos una referencia a la instancia de figura recién creada en la variable 'fig', y a partir de ella creamos una nueva instancia de 'axes' usando el método 'add_axes' en la instancia de clase 'Figura'

Losobjetos claves de Matplotlib.pyplot son :
1. Objeto figure
2. Objeto de axes

 Premisas Básicas

- Aprenda la terminología básica de matplotlib, específicamente qué es una figura y un eje.
- Utilice siempre la interfaz orientada a objetos. Acostúmbrese a usarlo desde el comienzo de su análisis.
- Lo más importante que debe recordar es que la Figura es la imagen final que puede contener 1 o más ejes. Los ejes representan una trama individual. Una vez que comprenda cuáles son y cómo acceder a ellos a través de la API orientada a objetos, el resto del proceso comienza a encajar.
- Programación orientada a objetos: use objetos a los que se puedan aplicar funciones y acciones, y ningún estado de objeto o programa debe ser global (como la API similar a MATLAB). Más conveniente cuando se crean varias figuras o cuando una figura contiene más de una subtrama. Trata a cada uno como un objeto.

```python
# Ejemplo básico

x = np.linspace(0, 5, 100)
y = x ** 3

# Se crea el objeto figura almacenado en la variable denomida Fig, inicialmente vacio
Fig = plt.figure()

# Esta función add_axes() toma una lista de valores. Los valores de izquierda a derecha son la izquierda, la parte inferior, el ancho y la altura.
axes = Fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('f(x)');
```

# Control de la apariencia de los ejes
Para configurar son los rangos de los ejes, se utiliza los métodos *'**set_ylim**'* y *'**set_xlim**'* en el objeto *'**axis**'*, o '***axis('tight')***' para obtener automáticamente rangos 'ajustados' de los ejes.

Se pueden activar y desactivar las líneas de cuadrícula a través del método ***'grid'*** del objeto del 'axis', además de personalizar la apariencia de las líneas.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

myaxes_1 = axes[0]
myaxes_2 = axes[1]

myaxes_1.plot(x, x**2, x, x**3)
myaxes_1.grid(True)
myaxes_1.set_title("RANGO POR DEFECTO")

myaxes_2.plot(x, x**2, x, x**3)
myaxes_2.set_ylim([0, 60])
myaxes_2.set_xlim([2, 5])
myaxes_2.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.2)
myaxes_2.set_title("RANGO MODIFICADO");
```

Se considera útil tener ejes dobles *x* o *y* en una figura. Más aún, al trazar curvas **con diferentes unidades juntas**. Matplotlib admite esto con las funciones ***twinx*** y ***twiny***.

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx() # sets the second y axis

ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volumen $(m^3)$", fontsize=18, color="red")

for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
def several_plots(ax,index, tamaño=15):
  location=['upper left','upper right','lower left','lower right']
  x=np.linspace(0,1,100)
  y=[x,x**2,x**3,x**4]
  Labels=['$y=x$','$y=x^{2}$','$y=x^{3}$','$y=x^{4}$']
  ax.plot(x,y[index],color="purple", lw=2,label=Labels[index])
  ax.set_xlim(0,1)
  ax.set_ylim(0,1)
  ax.legend(loc=location[index])
  ax.set_xlabel('$x$', fontsize=tamaño)
  ax.set_ylabel('$y$', fontsize=tamaño)
  ax.xaxis.set_major_formatter(plt.NullFormatter())
  ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Indicamos el número de filas y columnas (azúcar sintáctico)
nrows=ncols=2
# Retornamos el objeto figura y un arreglo de ejes
Myfig,Myaxes=plt.subplots(nrows,ncols)
plt.tight_layout()
index=0
for eje in Myaxes:
  for ax in eje:
    several_plots(ax,index, tamaño=20)
    index=index+1
# Para guardar las figuras en diferentes formatos
plt.savefig("plot_axes.png")
plt.savefig('plot_axes.pdf')
```

```python
x = np.linspace(0, 5, 60)
y = np.linspace(0, 5, 48)
def f(x, y):
    return np.sin(x)**5 + np.cos(y+17)**8
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
#pcolor
fig, ax = plt.subplots()
c = ax.pcolor(X, Y, Z, cmap='RdBu', vmin=abs(Z).min(), vmax=abs(Z).max())
ax.set_title('pcolor')
fig.colorbar(c, ax=ax);
```

```python
#imshow
fig, ax = plt.subplots()

im = ax.imshow(Z, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
im.set_interpolation('bilinear') # specific interpolation method set

cb = fig.colorbar(im, ax=ax)
```
